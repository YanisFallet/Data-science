```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')
plt.rcParams.update({
   "figure.facecolor":  (0.12 , 0.12, 0.12, 1),
    "axes.facecolor": (0.12 , 0.12, 0.12, 1),
})
```

```python
#création de la classe générale pour visualiser un graph
#OOP
class GraphVisualization: 
   
    def __init__(self): 
        self.visual = []
        
    def addEdge(self, a, b): 
        temp = [a, b] 
        self.visual.append(temp) 
          
    def visualize(self): 
        G = nx.Graph() 
        plt.figure(figsize=(12,8))
        G.add_edges_from(self.visual) 
        nx.draw_networkx(G) 
        plt.show()
```

```python
G = GraphVisualization() 
G.addEdge(0, 2) 
G.addEdge(1, 2) 
G.addEdge(1, 3) 
G.addEdge(5, 3) 
G.addEdge(3, 4) 
G.addEdge(1, 0) 
G.visualize() 
```

```python
#visualiser un graph à partir d'un matrice d'adjacence
#hérite des méthodes de la premiére classe

class Graphvisualisation_matrix(GraphVisualization):
    
    def __init__(self):
        self.visual=[]
        
    def visualize_matrix(self,adjacence_matrix):
        if adjacence_matrix.shape[0] != adjacence_matrix.shape[1]:
            raise ValueError
        n,m=adjacence_matrix.shape[0],adjacence_matrix.shape[1]
        for i in range(0,n):
            for j in range(0,m):
                d=adjacence_matrix[i,j]
                for k in range(d):
                    self.addEdge(i,j)
        self.visualize()       
```
